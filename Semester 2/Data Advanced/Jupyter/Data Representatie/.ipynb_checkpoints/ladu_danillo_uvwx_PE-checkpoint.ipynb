###  2. Genereer zelf de geboortedatum (willekeurige datum tussen 01,01,2011 en, 31,12,2011) 
###  3. Genereer de kolom inzet met 3 mogelijke categorieën

```python
import datetime
import random
import pandas

times = pandas.date_range(start='1/1/2011', end='31/12/2011', freq='D')

file = 'voetbal.xlsx'
df = pandas.read_excel(file, sheet_name='gegevens')
df.insert(8, 'geboortecatogorie', '0')

for index, item in df.iterrows():
    rtime = random.choice(times)
    if rtime >= datetime.datetime(2011, 10, 1, 0, 0):
        inzet = 'matig'
        geboortecat = 4
    elif rtime >= datetime.datetime(2011, 6, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 3
    elif rtime >= datetime.datetime(2011, 4, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 2
    else:
        inzet = 'zeer goed'
        geboortecat = 1

    df.at[index, ['inzet']] = inzet  
    df.at[index, ['geboortedatum']] = rtime.strftime('%d,%m,%Y')
    df.at[index, ['geboortecatogorie']] = geboortecat
    
    
print(print(df.head(5).to_string()))
```

Ik heb er voor gekozen deze datums op te slaan in "geboortedatum" en een extra kolom te maken "geboortecategorie" via [insert()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.insert.html#pandas.DataFrame.insert) met een standaard waarden van 0. Voor deze datum te generen heb ik een array laten generen tussen de twee datum met een interval van een dag [date_range()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html#pandas.date_range) hier ga ik dan telkens een willekeurig element uit nemen [choice()](https://docs.python.org/3.8/library/random.html#random.choice) dit functie komt uit de random class van python.  
Nu we dit hebben is het gewoon de waardes invullen in de tabellen, dit doe je door de [at()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.at.html#pandas.DataFrame.at) deze vraagt een index nummer en een kolom , in pandas kan je itereren over de rijen met een return van de element en de index [iterrows()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iterrows.html#pandas.DataFrame.iterrows) deze index gebruiken we voor de functie [at()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.at.html#pandas.DataFrame.at) en dan de uiteindelijke waarde die we meegeven.  
  
We hebben 4 categorieën van geboortes en 3 van inzet
  
|1|januari, februari, maart | zeer goed |
|2|april, mei, juni | goed |
|3|juli, augustus, september | goed |
|4|oktober, november, december | matig |
  
Om de categorie te bepalen tussen de twee datums 1-1-2011 en 31-12-2011 is het vrij simpel, is de datum groter of gelijk aan de eerste dag van oktober dan heb je categorie 4 zo niet doe je dit met juli en april ook, en als geen enkele voorwaarde voldoet dan is het categorie 1 , en dit dan voor elke element uit je iteratie, deze wordt ingevuld in je tabel. De [strftime()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.arrays.DatetimeArray.strftime.html?highlight=strftime#pandas.arrays.DatetimeArray.strftime) geeft een leesbaar datum terug.

###  5. Maak een staafdiagram van het aantal gemaakte goalen per positie. Deel de posities telkens ook nog op naar geboortecategorie en bespreek

```python
import datetime
import random
import pandas
import matplotlib.pyplot as plt

times = pandas.date_range(start='1/1/2011', end='31/12/2011', freq='D')

file = 'voetbal.xlsx'
df = pandas.read_excel(file, sheet_name='gegevens')
df.insert(8, 'geboortecatogorie', '0')

for index, item in df.iterrows():
    rtime = random.choice(times)
    inzet = ''
    geboortecat = 0
    if rtime > datetime.datetime(2011, 10, 1, 0, 0):
        inzet = 'matig'
        geboortecat = 4
    elif rtime > datetime.datetime(2011, 6, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 3
    elif rtime > datetime.datetime(2011, 4, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 2
    else:
        inzet = 'zeer goed'
        geboortecat = 1

    df.at[index, ['inzet']] = inzet  
    df.at[index, ['geboortedatum']] = rtime.strftime('%m,%d,%Y')
    df.at[index, ['geboortecatogorie']] = geboortecat

print(df.groupby(['positie'])['aantal gemaakte goalen'].sum().to_string())   
df.groupby(['positie'])['aantal gemaakte goalen'].sum().plot.bar(title='Aantal gemaakte goalen per positie', grid=True, figsize=(10, 10))
df.groupby(['positie', 'geboortecatogorie'])['aantal gemaakte goalen'].sum().unstack().plot.bar(title='Aantal gemaakte goalen per positie en geboortecatogorie', grid=True, figsize=(10, 10))
plt.show()
```

**Bespreek** : Uit deze kwantitatieve gegevens van aantal gemaakte doelpunten gegroepeerd op "positie" of "positie en geboortecategorie" zien we dat een keeper nooit scoort en dat een piloot het meeste scoort, linker en rechter vleugel doen het even goed. Bekijken we naar de lijn grafiek met geboorten zien we het zelfde alleen zijn de categorieën geregenereerd en kan ik niet concluderen welke geboorte categorie hier uitschiet aangezien dit telkens veranderd, de nadeel aan deze lijn grafiek is dat je de som van alle categorieën per positie moet tellen om te concluderen dat de linker en rechter vleugel hier het zelfde presteren. Dit maakt het moeilijker leesbaar.

###  6. Bereken het gemiddelde en de modus van kolom D (aantal gemaakte goalen) per positie

```python
import datetime
import random
import pandas

times = pandas.date_range(start='1/1/2011', end='31/12/2011', freq='D')

file = 'voetbal.xlsx'
df = pandas.read_excel(file, sheet_name='gegevens')
df.insert(8, 'geboortecatogorie', '0')

for index, item in df.iterrows():
    rtime = random.choice(times)
    inzet = ''
    geboortecat = 0
    if rtime > datetime.datetime(2011, 10, 1, 0, 0):
        inzet = 'matig'
        geboortecat = 4
    elif rtime > datetime.datetime(2011, 6, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 3
    elif rtime > datetime.datetime(2011, 4, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 2
    else:
        inzet = 'zeer goed'
        geboortecat = 1

    df.at[index, ['inzet']] = inzet  
    df.at[index, ['geboortedatum']] = rtime.strftime('%m,%d,%Y')
    df.at[index, ['geboortecatogorie']] = geboortecat
    

print('Gemiddelde op Aantal gemaakte goalen')
print('------------------------------------')
print(df.groupby(['positie'])['aantal gemaakte goalen'].mean().to_string())
print()
print('Modus op Aantal gemaakte goalen')
print('-------------------------------')
print(df.groupby(['positie'])['aantal gemaakte goalen'].median().to_string())
```

###  9. Maak een grafisch duidelijk hoe de verdeling is van de kolom inzet (vb cirkeldiagram met % ) en bespreek

```python
import datetime
import random
import pandas
import matplotlib.pyplot as plt

times = pandas.date_range(start='1/1/2011', end='31/12/2011', freq='D')

file = 'voetbal.xlsx'
df = pandas.read_excel(file, sheet_name='gegevens')
df.insert(8, 'geboortecatogorie', '0')

for index, item in df.iterrows():
    rtime = random.choice(times)
    inzet = ''
    geboortecat = 0
    if rtime > datetime.datetime(2011, 10, 1, 0, 0):
        inzet = 'matig'
        geboortecat = 4
    elif rtime > datetime.datetime(2011, 6, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 3
    elif rtime > datetime.datetime(2011, 4, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 2
    else:
        inzet = 'zeer goed'
        geboortecat = 1

    df.at[index, ['inzet']] = inzet  
    df.at[index, ['geboortedatum']] = rtime.strftime('%m,%d,%Y')
    df.at[index, ['geboortecatogorie']] = geboortecat
    
df.groupby(['inzet']).size().plot.pie(title='Circle Diagram Inzet', figsize=(10, 10))
plt.show()
```

###  10 Vergelijk de posities linkervleugel - rechtervleugel - piloot wat betreft het aantal gemaakte goalen mbv een boxplot en bespreek

```python
import datetime
import random
import pandas
import matplotlib.pyplot as plt

times = pandas.date_range(start='1/1/2011', end='31/12/2011', freq='D')

file = 'voetbal.xlsx'
df = pandas.read_excel(file, sheet_name='gegevens')
df.insert(8, 'geboortecatogorie', '0')

for index, item in df.iterrows():
    rtime = random.choice(times)
    inzet = ''
    geboortecat = 0
    if rtime > datetime.datetime(2011, 10, 1, 0, 0):
        inzet = 'matig'
        geboortecat = 4
    elif rtime > datetime.datetime(2011, 6, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 3
    elif rtime > datetime.datetime(2011, 4, 1, 0, 0):
        inzet = 'goed'
        geboortecat = 2
    else:
        inzet = 'zeer goed'
        geboortecat = 1

    df.at[index, ['inzet']] = inzet  
    df.at[index, ['geboortedatum']] = rtime.strftime('%m,%d,%Y')
    df.at[index, ['geboortecatogorie']] = geboortecat
    
df.boxplot(column=['aantal gemaakte goalen'], by='positie', figsize=(12, 12), grid=False)
plt.show()
```

***Bespreek***: Hier zien we de mediaan(modus)(Q2) dit is het groene lijn bij de vleugels staat deze op 1 dit zijn de waardes nog eens die we al eerder hebben berekend van de modus, dan heb je eerste kwartiel(Q1) en derde kwartiel(Q3) dit stelt dan de hele box voor en de twee uiteindes stellen min en max voor en het bolletjes zijn de uitschieters.  
Dit gezegd zien we dat de piloot weer het meeste scoort met een gemiddelde van 4 met een max van 6 en min van 1 , wat mij hier opvalt is da uitschieter op staart van 3.
