This notebook looks into a dataset of netflix movies and TV Shows and their directors, actors, duration and country.
We will be taking a look at:
- How many movies there are compared to TV Shows
- Releases per year (TV Shows and movies)
- Most popular directors in India
- Most popular directors in America
- Which TV Shows have the longest runtime
- What country creates the most content
- Most popular actors in India
- Most popular actors in the US

## Imports

```python
%config IPCompleter.greedy=True
import os
import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm

for dirname, _, filenames in os.walk('./'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

### We'll take a look at our available columns:

```python
content = pd.read_csv("netflix_titles.csv")
content.columns
```

#### Description of the columns:
<br>

| Feature       | Beschrijving                                                              |
|:--------------|:--------------------------------------------------------------------------|
| Type:         | Movie or TV Show?                                                         | 
| Title:        | Title of the movie or TV Show                                             |
| Director:     | Shows who directed the movie or TV Show                                   |
| Cast:         | Shows all actors that played in the movie or TV Show                      | 
| Country:      | Country that produced the movie or TV Show                                |
| Date_added:   | When it became available on Netflix                                       |
| Release_year: | Year of release                                                           |
| Rating:       | Motion picture content rating                                             |
| Duration:     | Duration of the movie in minutes or the TV Show in seasons                |
| Listed_in:    | Netflix categories it is listed under.                                    |
| Description:  | Short description of the movie or TV Show                                 |

<br style="clear:both" />

#### Next we'll take a look at our null values:

```python
#Kijken hoeveel van de data leeg is: 
content.isnull().sum()
```

The reason there's this many null values in the director column, is because TV Shows have a lot of different directors. This can vary per episode. Therefore it is less documented and this generates null values.

#### We'll fill all null values with an empty string so we don't have null values left. For the rating table, an empty value can even be used.

```python
content = content.fillna("")
content.isnull().sum()
```

```python
#variabel voor kleuren om te gebruiken door heel het notebook:
colors = ["red","blue","black","silver","gold",
         "orange","purple","gray","pink","green",
         "violet","brown","beige","turquoise","lime",
         "cyan", "yellow", "#23FF99", "#FFCBAD","#FF00FF"]

#Hoeveelheid films in vergelijking met series:
labels = 'MOVIES', 'TV SHOWS'
fig, plot = plt.subplots(figsize=(8,8))
plt.title('Hoeveelheid Films en Series')
plot.pie(content["type"].value_counts(), autopct="%1.0f%%", labels = labels, )
text="There are more movies than TV Shows on netflix \nThis is logical because a movie has a shorter duration. \nTV Shows have multiple seasons consisting of many episodes.\nA movie might have sequels but they count as separate movies and they're all up to a couple of hours."
plt.text(x=-1, y=-1.5, s=2,text=text, fontsize=14)
plt.show()
```

```python
#Hier zien we hoeveel films per jaar er released worden
#Ook zien we dat er elk jaar meer films released worden dan de vorige jaren.
filmsPerYear = pd.crosstab(content["release_year"], content["type"]).sort_values(by="release_year",ascending = False)[1:16] #We laten het jaar 2020 er tussen uit omdat dit jaar nog niet genoeg data heeft.
bar_plot = filmsPerYear.plot.barh(figsize=(12,10), color =colors)
bar_plot.set_title("Aantal releases per jaar")
bar_plot.set_ylabel("YEAR")
bar_plot.set_xlabel("RELEASES")
text = "We can clearly see that there's more movies than TV Shows every year. \nExcept for 2019, where the amount of movies released is exceptionally low.\n\nWe also notice that TV Shows are becoming a lot more popular. Their popularity grows faster than movies.\nMovies have also gained a lot of popularity over the years but peaked at 2017.\nAfter 2017 they started dropping in popularity, while TV Shows are still rising."
plt.text(x=0, y=-4, text=text, s=2, fontsize=14)

plt.show()
```

```python
#De director en het aantal gemaakte TV Shows
contentIndia = content[content["country"] == "India"]
bar_data = contentIndia['director'].value_counts()[1:11]
bar_plot = bar_data.plot.barh(figsize=(20, 12), color = colors)

bar_plot.set_title("Most popular directors in India")
bar_plot.set_xlabel("CONTENT DIRECTED")
bar_plot.set_ylabel("DIRECTOR")
plt.show()
```

```python
#De director en het aantal gemaakte Films
contentUS = content[content["country"] == "United States"]

bar_data = contentUS['director'].value_counts()[1:11]
bar_plot = bar_data.plot.barh(figsize=(20, 12), color = colors)
bar_plot.set_title("Most popular directors in America")
bar_plot.set_xlabel("CONTENT DIRECTED")
bar_plot.set_ylabel("DIRECTOR")
plt.show()
```

```python
#Create dataset with only Tv Shows
tvshows = content.loc[content["type"]=="TV Show",:]

#replace the duration with an integer for the number of seasons to sort correctly
#and rename the duration column
tvshows.loc[:,"duration"] = list(map(lambda x:int(x.rsplit(" ")[0]), tvshows["duration"]))

#Top 20 longest running Tv Shows
longestShows= tvshows.loc[:,["duration","title"]].sort_values(by="duration", ascending=False)[0:20]
labels = longestShows["title"].values

bar_plot = longestShows.plot.barh(figsize=(12,10), y = "duration", color=colors)
bar_plot.set_title("Tv Shows met de meeste seizoenen")
bar_plot.set_xlabel("SEASONS")
bar_plot.set_ylabel("TV SHOWS")

plt.yticks(ticks=range(0,20), labels=labels)
plt.legend("")
plt.show()
```

```python
#Landen met de meeste content gemaakt in hun land
top_countries = content.loc[content["country"]!="","country"].value_counts()[0:10]
pie_data = top_countries
pie_plot = pie_data.plot.pie(figsize=(12,12), autopct="%1.0f%%", rotatelabels = 1, labeldistance=0.65)
pie_plot.set_title("Content per land")
pie_plot.set_ylabel("")
text="Hollywood en Bollywood are the 2 largest producers.\nThis is why most content is created in India or America."
plt.text(x=-1,y=-1.2, s=2, fontsize=14, text=text)
plt.show()
```

```python
#Hoe vaak speelt een acteur mee in films/series (India)?

#Maak een list van actors
#Verander het type
#Filter en toon alleen de acteurs met 30+ netflix deelnames

#Het maken van een lijst
contentIndia = content.loc[content["country"]=="India",:]
casts = [];

#als het geen leeg veld is, splitten op komma en toevoegen
for cast in contentIndia["cast"]:
    if(cast!=""):
        casts.extend(str(cast).split(","))

casts2 = list(map(lambda x: x.strip(), casts))
unique, counts = np.unique(casts2, return_counts=True)
cast_count = np.asarray((unique, counts)).T
summary_cast = pd.DataFrame(cast_count)

#Veranderen van het datatype om te kunnen sorteren
summary_cast[1] = summary_cast[1].astype(int)

#Het filteren en sorteren van de summary_cast
summary_cast.columns=["actor","plays"]
summary_cast = summary_cast.sort_values(by="plays", ascending=False)[0:10]

#plot
bar_data = summary_cast
bar_plot = bar_data.plot.barh(figsize=(10, 12), x=0, y=1, color=colors)
plt.ylabel('Counts')
plt.xlabel('Actors')
plt.legend("")
plt.title('Most popular actors in India')
plt.show()
```

```python
#Hoe vaak speelt een acteur mee in films/series (Amerika)?

#Maak een list van actors
#Verander het type
#Filter en toon alleen de acteurs met 30+ netflix deelnames

#Het maken van een lijst
contentUS = content.loc[content["country"]=="United States",:]
casts = [];
for cast in contentUS["cast"]:
    if(cast!=""):
        casts.extend(str(cast).split(","))

casts2 = list(map(lambda x: x.strip(), casts))
unique, counts = np.unique(casts2, return_counts=True)
cast_count = np.asarray((unique, counts)).T
summary_cast = pd.DataFrame(cast_count)
#Veranderen van de types
summary_cast[1] = summary_cast[1].astype(int)

#Het filteren en sorteren van de summary_cast
summary_cast.columns=["actor","plays"]
summary_cast = summary_cast.sort_values(by="plays", ascending=False)[0:10]

#plot
bar_data = summary_cast
bar_plot = bar_data.plot.barh(figsize=(10, 12), x=0, y=1, color=colors)
plt.ylabel('Counts')
plt.xlabel('Actors')
plt.legend("")
plt.title('Most popular actors in America')
text="Voice actors are also actors, so not all actors in this top 10 are actually popular amongst the general public.\nFred and Erin for example are voice actors."
plt.text(x=0,y=-2, s=2, fontsize=14, text=text)
plt.show()
```

```python
#Rating van netflix
summary_rating = (content.groupby('rating').size().to_frame('count').reset_index())

#sorteren op count
summary_rating.sort_values(by=['count'], inplace=True)

#plot
bar_data = summary_rating
bar_plot = bar_data.plot.barh(figsize=(12, 12), x=0, y=1, color=colors)
plt.ylabel('Rating')
plt.xlabel('Count')
plt.title('Count per rating')
plt.legend("")
text= "We notice that the most popular target audience are adults and secondly teenagers older than 14.\nTV-MA is a rating for mature audiences, so adults.\nTV-14 means that it might not be suitable for children below the age of 14."
plt.text(x=0,y=-2.5, s=2, fontsize=14, text=text)
plt.show()
```
